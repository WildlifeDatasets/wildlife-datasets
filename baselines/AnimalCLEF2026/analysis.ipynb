```python
import os
import sys

sys.path.append("../..")

import matplotlib.pyplot as plt
import numpy as np

from wildlife_datasets.datasets import AnimalCLEF2026
from wildlife_datasets.splits import analyze_split
```

```python
root = "/data/wildlife_datasets/data/AnimalCLEF2026"
dataset = AnimalCLEF2026(root)
```

```python
img_max = 200
n_save = 0

sizes = {
    "LynxID2025": (img_max, img_max),
    "SalamanderID2025": (0.4 * img_max, img_max),
    "SeaTurtleID2022": (img_max, img_max),
    "TexasHornedLizards": (0.7 * img_max, img_max),
}
os.makedirs("figures", exist_ok=True)
for name, metadata_subset in dataset.metadata.groupby("dataset"):
    size = np.round(sizes[name]).astype(int)
    dataset_subset = dataset.get_subset(metadata_subset.index)
    dataset_subset.set_transform(lambda x: x.resize(size))
    for i in range(n_save):
        qwe = dataset_subset.plot_grid(
            n_rows=3,
            n_cols=2,
            img_min=np.min(size),
            rotate=False,
            keep_transform=True,
            background_color=(255, 255, 255),
        )
        plt.savefig(f"figures/{name}_{i}.png", dpi=600, bbox_inches="tight")
        plt.close()
```

```python
for name, metadata_subset in dataset.metadata.groupby("dataset"):
    print(name)
    idx_train = metadata_subset[metadata_subset["split"] == "train"].index
    idx_test = metadata_subset[metadata_subset["split"] == "test"].index
    analyze_split(metadata_subset, idx_train, idx_test)
    print("\n\n")
```
